```python
# Takes in the cropped 3d images and adds padding to make them all the same size 
def pad(images, maxes):

    zf_max, yf_max, xf_max = (maxes[2], maxes[1], maxes[0]) # The destination size of dimensions
    
    z_diff, y_diff, x_diff = zf_max - images.shape[2], yf_max - images.shape[1], xf_max - images.shape[0]
    
    print("Differences:", x_diff, y_diff, z_diff)
    
    #pad_stack = np.empty((images.shape[2], images.shape[1], images.shape[0]))
       
    z_odd = z_diff%2
    y_odd = y_diff%2
    x_odd = x_diff%2

    # Pad all the dimensions as equally as possible on each side (6 sides in a cube)
    if z_odd:
        z_pad = (int((z_diff - 1) / 2)+1, (int((z_diff - 1) / 2)))
    else:
        z_pad = (int(z_diff / 2), int(z_diff / 2)) 

    if y_odd:
        y_pad = (int((y_diff - 1) / 2)+1, (int((y_diff - 1) / 2)))
    else:
        y_pad = (int(y_diff / 2), int(y_diff / 2))                 

    if x_odd:
        x_pad = (int((x_diff - 1) / 2)+1, (int((x_diff - 1) / 2)))
    else:
        x_pad = (int(x_diff / 2), int(x_diff / 2))

    images_padded = np.pad(images, [x_pad, y_pad, z_pad], mode='constant')

    return images_padded
```
