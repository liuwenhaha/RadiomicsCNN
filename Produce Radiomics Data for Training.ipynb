```python
data_dir = 'D:\Data\MAASTRO-CT-Lung1(nrrds)'
folders  = os.listdir(data_dir)
folders  = [folder for folder in folders if folder[:5] == "LUNG1"]
folders  = [os.path.join(data_dir, folder) for folder in folders]
```

```python
img_path = os.path.join(folders[0], 'image.nrrd')
msk_path = os.path.join(folders[0], 'mask.nrrd')
img, img_header = nrrd.read(img_path)
msk, msk_header = nrrd.read(msk_path)
img.shape
```

```python
params = os.path.join(data_dir, "Params.yaml")
```

```python
extractor = featureextractor.RadiomicsFeatureExtractor(params)
result = extractor.execute(img_path, msk_path)
result_dict = {}
keys = []
for key, val in six.iteritems(result):
    #print("\t%s: %s" %(key, val))
    keys.append(key)
    result_dict[key] = val
```

```python
df = pd.DataFrame(columns = keys)
```

```python
for n, folder in enumerate(folders):
    #if n > 5: break
    print("Processing:", folder)
    img_path = os.path.join(folder, 'image.nrrd')
    msk_path = os.path.join(folder, 'mask.nrrd')
    extractor = featureextractor.RadiomicsFeatureExtractor(params)
    result = extractor.execute(img_path, msk_path)
    result_dict = {}
    for key, val in six.iteritems(result):
        #print("\t%s: %s" %(key, val))
        result_dict[key] = val
    df = df.append(result_dict, ignore_index = True)
```

```python
df.to_csv('radiomics_features.csv')
```
